## Gráficos das análises: `Sequencial.cpp`



```python
import matplotlib.pyplot as plt

# Tamanhos das matrizes:
sizes = [100, 200, 400, 800, 1600, 3200, 6400]

# Tempos correspondentes em milissegundos:
times = [0.88256, 6.6976, 43.0316, 413.927, 8285.05, 102822, 10743000000]

plt.figure(figsize=(7,5))
plt.plot(sizes, times, marker='o', linestyle='-', color='green')
plt.title('Tempo de cálculo x Tamanho da matriz (escala log-log)')
plt.xlabel('Tamanho da Matriz (M x N)')
plt.ylabel('Tempo (ms)')
plt.xscale('log')
plt.yscale('log')
plt.grid(True, which="both", ls="--")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Tamanho da matriz e tempo médio:
sizes = [400]
mean_times = [43.226]

plt.figure(figsize=(6,5))

# Ponto destacado com cor e tamanho maior:
plt.scatter(sizes, mean_times, color='#ff7f0e', s=200, edgecolor='black', zorder=5)

# Linha horizontal para referência:
plt.axhline(y=mean_times[0], color='gray', linestyle='--', linewidth=1, alpha=0.7)

# Anotação do valor:
plt.text(sizes[0], mean_times[0]*1.05, f'{mean_times[0]:.3f} ms', ha='center', fontsize=12, fontweight='bold')

# Configurações do gráfico:
plt.title('Tempo Médio de Execução - Matriz 400x400', fontsize=16, fontweight='bold')
plt.xlabel('Tamanho da Matriz (M x N)', fontsize=12)
plt.ylabel('Tempo Médio (ms)', fontsize=12)
plt.ylim(0, mean_times[0]*1.5)
plt.xticks(sizes)  # mostra apenas o ponto 400.
plt.grid(True, linestyle='--', alpha=0.7, zorder=0)

plt.show()
```

## Experimentos

### E1
## Tempo médio X Tamanho da matriz

```python
import matplotlib.pyplot as plt
import numpy as np

# --- DADOS SEQUENCIAL ---
sizes_seq = [100, 200, 400, 800, 1600, 3200, 6400]
times_seq = [0.88256, 6.6976, 43.0316, 413.927, 8285.05, 102822, 10743000000000]

# --- DADOS THREADS ---
sizes_thr = [100, 200, 400, 800, 1600, 3200, 6400]
times_thr = [0.88256, 3.2365, 37.0316, 312.927, 5958.05, 92822, 127593156]

# --- DADOS PROCESSOS ---
sizes_proc = [100, 200, 400, 800, 1600, 3200, 6400]
times_proc = [0.82562, 3.6749, 41.0326, 332.972, 6285.05, 112822, 1074300000]

# --- TEMPOS MÉDIOS ---
# Sequencial - matriz 100x100
sizes_seq_mean = [400]
mean_seq = [43.226]

# Threads - matriz 100x100
times_thr_mean_list = [0.463542, 0.340781, 0.494641, 0.448143,
                       0.51054, 0.668898, 0.202982, 0.363203]
sizes_thr_mean = [100]
mean_thr = [np.mean(times_thr_mean_list)]

# Processos - matriz 100x100
times_proc_mean_list = [0.348515, 0.306426, 0.171973, 0.170791, 0.508507, 0.235824, 0.290907]
sizes_proc_mean = [100]
mean_proc = [np.mean(times_proc_mean_list)]

# --- PLOTAGEM ---
plt.figure(figsize=(8,6))

# Curvas principais
plt.plot(sizes_seq, times_seq, marker='o', linestyle='-', color='green', label="Sequencial")
plt.plot(sizes_thr, times_thr, marker='o', linestyle='-', color='orange', label="Threads")
plt.plot(sizes_proc, times_proc, marker='o', linestyle='-', color='blue', label="Processos")

# Médias destacadas
plt.scatter(sizes_seq_mean, mean_seq, color='green', s=180, edgecolor='black', zorder=5, label="Média Sequencial (100x100)")
plt.scatter(sizes_thr_mean, mean_thr, color='orange', s=180, edgecolor='black', zorder=5, label="Média Threads (100x100)")
plt.scatter(sizes_proc_mean, mean_proc, color='blue', s=180, edgecolor='black', zorder=5, label="Média Processos (100x100)")

# Configurações do gráfico
plt.title('Comparação: Tempo Médio x Tamanho da Matriz', fontsize=14, fontweight='bold')
plt.xlabel('Tamanho da Matriz (M x N)', fontsize=12)
plt.ylabel('Tempo (ms)', fontsize=12)
plt.xscale('log')
plt.yscale('log')
plt.grid(True, which="both", ls="--", alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```
